```python
#load

run = np.load('running_12and100.npz')
Val_c12=run['Val_c12']
FEV_c12=run['FEV_c12']
Val_c100=run['Val_c100']
FEV_c100=run['FEV_c100']
Val_c1000=run['Val_c1000']
FEV_c1000=run['FEV_c1000']

finished = 5
```

```python
# selecting best
#rep,Data,reg,N,model/mean
Data = 2**np.arange(9,17)
D=len(Data)

#best reg?
ind_r12 = np.zeros([finished,D,12])
ind_r100 = np.zeros([finished,D,100])
ind_r1000 = np.zeros([finished,D,1000])

ind_m12 = np.zeros([finished,D,3,12])
ind_m100 = np.zeros([finished,D,3,100])
ind_m1000 = np.zeros([finished,D,3,1000])

fev_corrected12 = np.zeros([finished,D,12])
fev_corrected100 = np.zeros([finished,D,100])
fev_corrected1000 = np.zeros([finished,D,1000])

Neurons=[12,100,1000]
for rep in range(finished):
    for n in range(3):
        N=Neurons[n]
        for d in range(D):
            #best reg?
            for i in range(N):
                if n==0:
                    ind_r12[rep,d,i] = np.argmin(np.min(Val_c12[rep,d,:,i,:],-1))
                if n==1:
                    ind_r100[rep,d,i] = np.argmin(np.min(Val_c100[rep,d,:,i,:],-1))
                if n==2:
                    ind_r1000[rep,d,i] = np.argmin(np.min(Val_c1000[rep,d,:,i,:],-1))
                #best model? or mean?
                for r in range(3):
                    if n==0:
                        ind_m12[rep,d,r,i] = np.argmin(Val_c12[rep,d,r,i,:])
                    if n==1:
                        ind_m100[rep,d,r,i] = np.argmin(Val_c100[rep,d,r,i,:])
                    if n==2:
                        ind_m1000[rep,d,r,i] = np.argmin(Val_c1000[rep,d,r,i,:])
                
                #choose best reg and model/mean for all units and reps
                if n==0:
                    #best reg and model/mean
                    fev_corrected12[rep,d,i] = FEV_c12[rep,d,
                                           np.int(ind_r12[rep,d,i]),i,
                                           np.int(ind_m12[rep,d,np.int(ind_r12[rep,d,i]),i])]
                if n==1:
                    fev_corrected100[rep,d,i] = FEV_c100[rep,d,
                                           np.int(ind_r100[rep,d,i]),i,
                                           np.int(ind_m100[rep,d,np.int(ind_r100[rep,d,i]),i])]
                if n==2:
                    fev_corrected1000[rep,d,i] = FEV_c1000[rep,d,
                                           np.int(ind_r1000[rep,d,i]),i,
                                           np.int(ind_m1000[rep,d,np.int(ind_r1000[rep,d,i]),i])]
                    
#now look at performance over repetitions
for n in range(3):
    N=2**(n+2)*64
    fev_m = np.zeros(D)
    fev_sd = np.zeros(D)
    for d in range(D):
        if n==0:
            fev_m[d] = np.mean(np.mean(fev_corrected12[:,d,:],-1))#mean over reps for each neuron, then mean over neurons
            fev_sd[d] = np.std(np.mean(fev_corrected12[:,d,:],-1))#mean over reps for each neuron, then std over neurons
        if n==1:
            fev_m[d] = np.mean(np.mean(fev_corrected100[:,d,:],-1))#mean over reps for each neuron, then mean over neurons
            fev_sd[d] = np.std(np.mean(fev_corrected100[:,d,:],-1))#mean over reps for each neuron, then std over neurons
        if n==2:
            fev_m[d] = np.mean(np.mean(fev_corrected1000[:,d,:],-1))#mean over reps for each neuron, then mean over neurons
            fev_sd[d] = np.std(np.mean(fev_corrected1000[:,d,:],-1))#mean over reps for each neuron, then std over neurons
    
    plt.errorbar(Data,fev_m,fev_sd)
plt.gca().set_xscale("log", nonposx='clip')
plt.legend(Neurons)
plt.show()
```
